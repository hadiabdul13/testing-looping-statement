# Enhancing Memory and Time Efficiency for Looping Statements in Python

In this section, we will evaluate the memory usage and time efficiency of looping statements using for and while loops, as well as yield as a generator function for both looping methods.

Looping is a fundamental concept in programming, especially for beginners learning Python. It allows repeated execution of a set of instructions until a condition is met, making automation and data processing easier, particularly with large datasets.

# Memory Efficiency: For vs While vs Yield

Loops using for and while store all iteration results in memory, consuming significant space. However, yield, a feature in generator functions, generates values one at a time, reducing memory usage significantly.


# Time Efficiency Comparison

Using Pythonâ€™s time module, the execution times were tested. While for and while loops have similar memory usage, for loops execute slightly faster due to shorter syntax. However, yield is the most efficient, both in time and memory.

# Conclusion

- For and while loops consume a lot of memory since they store all values.
- While loops are more flexible but slightly slower than for loops.
- Yield (generator) is the best choice for large-scale data processing due to its minimal memory usage and faster execution.

Understanding these differences helps programmers choose the most efficient method based on memory and time constraints.

```python
import pandas as pd
import matplotlib.pyplot as plt
import sys
import time
```

```python
# Structure and simple syntax of using for

for i in range(5):
    print(i)
```

```python
# Structure and simple syntax of using while

i = 0
while i < 5:
    print(i)
    i += 1
```

```python
# Structure and simple syntax of using yield

def generator():
  for i in range(5):
    yield i

for value in generator():
  print(value)
```

```python
# Function using a for loop to create and return a list of numbers from 0 to n-1
def for_loop(n):
    return [i for i in range(n)]  # List comprehension to generate numbers

# Function using a generator to yield numbers one by one instead of storing them in memory
def generator_loop(n):
    for i in range(n):
        yield i  # Yield returns one value at a time, saving memory

# Function using a while loop to create and return a list of numbers from 0 to n-1
def while_loop(n):
    i = 0
    result = []
    while i < n:
        result.append(i)  # Append numbers to the list
        i += 1
    return result  # Return the complete list

n = 1000000  # Define the size of the range

# Measure and print memory usage of each function
print("For loop memory usage:", sys.getsizeof(for_loop(n)), "bytes")
print("Yield (generator) memory usage:", sys.getsizeof(generator_loop(n)), "bytes")
print("While loop memory usage:", sys.getsizeof(while_loop(n)), "bytes")
```

```python
# Creating a DataFrame for memory usage analysis
data_memory = {
    "Loop Type": ["For Loop", "While Loop", "Yield (Generator)"],
    "Memory Usage (bytes)": [8448728, 8448728, 200]  # Updated memory values
}

df_memory = pd.DataFrame(data_memory)

# Data frame untuk penggunaan memori untuk setiap loop
print(df_memory)

# Plotting bar chart using matplotlib
plt.figure(figsize=(8, 5))
plt.bar(df_memory["Loop Type"], df_memory["Memory Usage (bytes)"], color=["blue", "green", "red"])
plt.xlabel("Loop Type")
plt.ylabel("Memory Usage (bytes)")
plt.title("Memory Usage Comparison: For Loop vs While Loop vs Yield")
plt.show()
```

```python
# Testing for execution time
n = 1000000

# Measuring execution time for for loop
start = time.time()
for_loop(n)
time_for = time.time() - start

# Measuring execution time for while loop
start = time.time()
while_loop(n)
time_while = time.time() - start

# Measuring execution time for yield (generator)
start = time.time()
for _ in generator_loop(n):
    pass
time_yield = time.time() - start
```

```python
# Creating a DataFrame for time comparison
data_time = {
    "Loop Type": ["For Loop", "While Loop", "Yield (Generator)"],
    "Execution Time (seconds)": [time_for, time_while, time_yield]
}

df_time = pd.DataFrame(data_time)

# Display DataFrame
print(df_time)

# Plotting bar chart for execution time comparison
plt.figure(figsize=(8, 5))
plt.bar(df_time["Loop Type"], df_time["Execution Time (seconds)"], color=["blue", "green", "red"])
plt.xlabel("Loop Type")
plt.ylabel("Execution Time (seconds)")
plt.title("Execution Time Comparison: For Loop vs While Loop vs Yield")
plt.show()
```

```python
# Measuring execution time for for loop
n = 1000000  # Example with 1 million iterations

start = time.time()
for_loop(n)
time_for = time.time() - start

# Print the execution time
print(start)
print(time_for)
```

```python
# Analysing execution time

n = 1000000

start = time.time()
for_loop(n)
time_for = time.time() - start

start = time.time()
while_loop(n)
time_while = time.time() - start

start = time.time()
for _ in generator_loop(n):
    pass
time_yield = time.time() - start

# Creating DataFrame
data_time = {
    "Loop Type": ["For Loop", "While Loop", "Yield (Generator)"],
    "Execution Time (seconds)": [time_for, time_while, time_yield]
}

df_time = pd.DataFrame(data_time)

# Displaying DataFrame
print(df_time)

# Visualisation with bar chart
plt.figure(figsize=(8, 5))
plt.bar(df_time["Loop Type"], df_time["Execution Time (seconds)"], color=["blue", "green", "red"])
plt.xlabel("Loop Type")
plt.ylabel("Execution Time (seconds)")
plt.title("Execution Time Comparison: For Loop vs While Loop vs Yield")
plt.show()
```
